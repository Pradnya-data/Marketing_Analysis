# Univariate Analysis


```python
# Campaign Channel
df['CampaignChannel'].value_counts()
```

```python
#Campaign Type
df['CampaignType'].value_counts()
```

```python
df['ConversionRate'].max()
```

Age Distribution


```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Income Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Income'], bins=30, kde=True, color='salmon')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

```python
# Gender Distribution
msc=df['Gender'].value_counts()
print(msc.index)
plt.pie(msc,labels=msc.index,autopct="%d%%")
plt.title('Marital Status')
plt.ylabel('')
plt.show()
```

```python
# Campaign Channel Success

sns.countplot(x='CampaignChannel', data=df)
plt.title('Conversion Count by Campaign Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
# # Campaign Type Success

sns.countplot(x='CampaignType',hue='Gender', data=df)
plt.title('Conversion Count by Campaign Channel')
plt.xticks(rotation=45)
plt.show()
```

# Bivariate Analysis

Click through rate and campaign type


```python
sns.boxplot(x='CampaignType', y='ClickThroughRate', data=df)
plt.title('Click Through Rate by Campaign Type')
plt.xticks(rotation=45)
```

```python
# income vs AdSpend
sns.scatterplot(x='Income', y='AdSpend',hue='Conversion', data=df)
plt.title('Ad Spend vs Income (Colored by Conversion)')
```

Reviewing Email Clicks and Emails Opened

```python
plt.figure(figsize=(4,3))
sns.histplot(df['EmailOpens'], kde=True, color='teal')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.histplot(df['EmailClicks'], kde=True, color='purple')
plt.show()
```
